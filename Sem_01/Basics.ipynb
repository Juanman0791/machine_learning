### Let's play with the data

```
# Printing the image shape (size of the image)
img.shape
```

```
# Looking through the image...
pixel_value = img[100,150]
print("Pixel value at [100,150]:", pixel_value)
# print among image indexes
print("Example: print pixel values from row 100 to 105 and column 150 to 155")
for row in range(100,105):
  for col in range(150,155):
    print(f"Pixel Value are[{row},{col}]:{img[row,col]}")
```

```
# Changing its channels values!

img_color = cv2.imread('/content/drive/MyDrive/machine learnig/machine_learning/Images/Lenna_(test_image).png')

# Cambiar los valores del canal azul
img_color[:, :, 0] = 0

# Cambiar los valores del canal verde
img_color[:, :, 1] = img_color[:, :, 1] // 2

# Cambiar los valores del canal rojo
img_color[:, :, 2] = 255
# Mostrar la imagen modificada
cv2_imshow(img_color)
```

```
# Changing its channels values!
img = cv2.imread('/content/drive/MyDrive/machine learnig/machine_learning/Images/Lenna_(test_image).png')
img[:, :, 1] = 255  # Cambiar el canal verde
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convertir BGR a RGB para mostrar correctamente
plt.show()  # Voil!
```

### Load an image in grayscale

```
# Loading a color image in grayscale
img_gray = cv2.imread('/content/drive/MyDrive/machine learnig/machine_learning/Images/Lenna_(test_image).png')
Gris=cv2.cvtColor(img_gray, cv2.COLOR_BGR2GRAY)
cv2_imshow(Gris)
```

### Play with the image


```
# Printing the image shape
print(Gris.shape)
```

```
# Regarding through the image... Again

# Acceder a un pxel especfico en [50, 75]
pixel_value = Gris[50, 75]
print("Pixel value at [50,75]:", pixel_value)

# Imprimir los valores de pxeles en un rango especfico
print("Example: print pixel values from row 50 to 55 and column 75 to 80")
for row in range(50, 55):
    for col in range(75, 80):
        print(f"Pixel Value at [{row},{col}]: {Gris[row, col]}")
```

### Going further

```
# Messing everything around.
img = cv2.imread('/content/drive/MyDrive/machine learnig/machine_learning/Images/Lenna_(test_image).png')
img[:, :, 0] = 0
img[:, :, 1] = img[:, :, 1] // 2  # Reducir el canal verde a la mitad
img[:, :, 2] = 255  # Establecer el canal rojo al mximo
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convertir BGR a RGB para mostrar correctamente
plt.show()
```

```
# Playing with image intensities
img_gray = cv2.imread('/content/drive/MyDrive/machine learnig/machine_learning/Images/Lenna_(test_image).png', cv2.IMREAD_GRAYSCALE)

for r in range(img_gray.shape[0]):  # Go through rows
    for c in range(img_gray.shape[1]):  # Go through cols
        if img_gray[r, c] < 128:  # Create a simple condition
            img_gray[r, c] = 255  # Set pixel to white if below threshold

plt.imshow(img_gray, cmap='gray')  # Show image with grayscale colormap
plt.show()
```
